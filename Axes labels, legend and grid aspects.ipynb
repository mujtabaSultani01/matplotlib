## Axes labels, Legend & Grid aspects of 2D Chart
**Axes:** An axis is a line drawn on a chart, graph, or diagram that indicates a point of origin for the measurement of values. It is used to measure values on two scales, a vertical (y-axis) and a horizontal (x-axis) axis.

**Legend:** A legend is a visual explanation of the symbols used in a chart or graph. It is used to explain what each line, bar, or other symbol represents in the graph, making the data easier to interpret.

**Grid:** A grid is a collection of horizontal and vertical lines used to divide a chart or graph into sections. It is used to divide a graph into equal parts, making it easier to interpret and compare data.

```python
# To import the library:
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# So here we use some static data for 2D plot.
# Here are four lists which represent the weather data. first list containt the days of the week. 2nd list contain the
# maximum temperatures recorded on the specific days. 3rd list contain minimum temperature recorded on each day. 4th list
# contain the average data recorded on each day.
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# To plot the lists, we call to plot fucntion and pass the x-axis and y-axis. 
# So we plot the chart for my maximum temperature:
plt.plot(days, max_t)
```

```python
# In a same chart (all in one chart) we can plot three different lines for each type of weathers:
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# To know which one is temperature and which one is week days, we set chart labels with addition to the chart title:
plt.title("Weather Chart")
plt.xlabel("Week days")
plt.ylabel("Temperature")

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
# Now again we don't know which line is representing max, min or average weather? So for that we can use legend. in order to
# have a legend, we need to define the label to each of the line:
plt.title("Weather Chart")
plt.xlabel("Week days")
plt.ylabel("Temperature")

plt.plot(days, max_t, label = "Max")
plt.plot(days, min_t, label = "Min")
plt.plot(days, avg_t, label = "Avg")

# Now we can set the legend property:
plt.legend()
```

```python
# To change the location of the legend, we can use 'loc' property:
plt.title("Weather Chart")
plt.xlabel("Week days")
plt.ylabel("Temperature")

plt.plot(days, max_t, label = "Max")
plt.plot(days, min_t, label = "Min")
plt.plot(days, avg_t, label = "Avg")

# Now we can set the legend property:
plt.legend(loc = "upper right")
```

```python
# By default the 'loc' property is set to 'best', it means model will decide where to locate the legend according to free space as we see in [8].
# If we change the first value of max temperature from 50 to 34, so the location of legend will change automatically:
plt.title("Weather Chart")
plt.xlabel("Week days")
plt.ylabel("Temperature")

max_te=[34,51,52,48,47,49,46]

plt.plot(days, max_te, label = "Max")
plt.plot(days, min_t, label = "Min")
plt.plot(days, avg_t, label = "Avg")

# Now we can set the legend property:
plt.legend()
```

```python
# You can also add 'shadow' and 'fontsize' properties:
plt.title("Weather Chart")
plt.xlabel("Week days")
plt.ylabel("Temperature")

max_te=[34,51,52,48,47,49,46]

plt.plot(days, max_te, label = "Max")
plt.plot(days, min_t, label = "Min")
plt.plot(days, avg_t, label = "Avg")

# Now we can set the legend property:
plt.legend(shadow = True, fontsize = "small")
```

```python
# Another thing that people normally interested with in 2D plotting is grid. right now the convas of the upper 2D plot is 
# blank, sometime we see their is a grid for better data visualization. So for that there is a property 'grid'. 
plt.title("Weather Chart")
plt.xlabel("Week days")
plt.ylabel("Temperature")

max_te=[34,51,52,48,47,49,46]

plt.plot(days, max_te, label = "Max")
plt.plot(days, min_t, label = "Min")
plt.plot(days, avg_t, label = "Avg")

# Now we can set the legend property:
plt.legend(shadow = True, fontsize = "small")
plt.grid()
```
