## Format Strings 
**Format strings** in the matplotlib library are strings used to define the appearance of plot elements such as lines, markers, and text. These strings are composed of a set of formatting characters that allow you to control the look of individual elements such as color, line style, and marker size.

```python
# Let's first import the library:
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# We use some static datat to plot the chart.  
# Weather data:
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
# To plot the data we simple call to library and supply the x and y arguments:
plt.xlabel("Week days")
plt.ylabel("Temperature")
plt.title("Weather Chart")
plt.plot(x,y)
```

```python
# Now we can control visual properties of this chart by specifying format strings as 3rd argument. for example we want 
# change color to 'blue', we set the marker to '+' and we set the line to 'dashed'. So the way we do it is:
plt.xlabel("Week days")
plt.ylabel("Temperature")
plt.title("Weather Chart")
plt.plot(x, y, "b+--")
# here b => blue, + => marker, and -- => dashed 
# The order is not an issue (doesn't matter), it will work for you.
```

* You can see the list of all format strings in the matplotlib library official website.

```python
# Or:
plt.xlabel("Week days")
plt.ylabel("Temperature")
plt.title("Weather Chart")
plt.plot(x, y, "gD--")
```

```python
# You can plot the same chart using keyword arguments instead of format strings:
plt.xlabel("Week days")
plt.ylabel("Temperature")
plt.title("Weather Chart")
plt.plot(x, y, color = "green", marker = "D", linestyle = "dashed")
```

```python
# To control the marker size, then there is 'markersize' argument that we can use:
plt.xlabel("Week days")
plt.ylabel("Temperature")
plt.title("Weather Chart")
plt.plot(x, y, color = "Maroon", marker = "*", linestyle = "dotted", markersize = 10)
```

```python
# The next property is called alpha. Alpha basically control the transferency. The range is between 0 and 1.
# See for range 0.8:
plt.xlabel("Week days")
plt.ylabel("Temperature")
plt.title("Weather Chart")
plt.plot(x, y, color = "Maroon", marker = "*", linestyle = "dotted", markersize = 10, alpha = 1)
```

```python
# Now see 'alpha' for 0:
plt.xlabel("Week days")
plt.ylabel("Temperature")
plt.title("Weather Chart")
plt.plot(x, y, color = "Maroon", marker = "*", linestyle = "dotted", markersize = 10, alpha = 0.2)
```

* There are many properties that plot API support, you can check them in the matplotlib officail website.
