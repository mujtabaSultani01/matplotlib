## Bar Chart in Matplotlib
A **bar** chart in matplotlib is a type of chart that displays data using rectangular bars with lengths proportional to the values that they represent. Bar charts are used to compare values between different categories. They can be used to show the distribution of data over time, or compare the differences between different groups.

```python
# Let's import matplotlib library:
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Let's have some static revenue data for some US companies in the following two lists.
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
# So now to create a bar chart for them, we call to bar function and pass the values. Here we first change the company list
# into numpy array and then pass it to bar function (we can also use company list):
com = np.arange(len(company))
com
```

```python
# Now we pass the the values:
plt.bar(com, revenue)
```

```python
# Now to use labels instead of 'com' numbers, we use 'xticks' function. Ti taks two arguments: 1st is 'com' and the 2nd one 
# is the company list.
plt.xticks(com, company)
plt.bar(com, revenue)
```

```python
# Let's add labels to the chart:
plt.xticks(com, company)
plt.title("Companies Revenue")
plt.ylabel("revenue ($)")
plt.bar(com, revenue, label = "revenue")
plt.legend()
```

```python
# Now if we want to plot two different measures on the chart? if we have profit numbers for each of the company:
profit=[40,2,34,12]
```

```python
# To plot the bar chart for both revenue and profits:
plt.xticks(com, company)
plt.title("Companies Revenue")
plt.ylabel("revenue ($)")
plt.bar(com, revenue, label = "revenue")
plt.bar(com, profit, label = "profit")
plt.legend()
```

```python
# Now as we see both the revenue and profit is plotting but it's plotting in a same chart, the same bar containing both 
# revenue and profit. We want to generate sides by sides bar for each of them. The reason why they both plotted in singles 
# bars is, we have the same 'com' for both of them. Now if we make them a bit different by using subtracting some numbers, 
# then we will get almost good chart but it's not perfect. And that's why we created numpy array for 'company' list.
plt.xticks(com, company)
plt.title("Companies Revenue")
plt.ylabel("revenue ($)")
plt.bar(com - 0.2, revenue, label = "revenue")
plt.bar(com + 0.2, profit, label = "profit")
plt.legend()
```

```python
# To control the width of bars, we use width property:
plt.xticks(com, company)
plt.title("Companies Revenue")
plt.ylabel("revenue ($)")
plt.bar(com - 0.2, revenue, width = 0.4, label = "revenue")
plt.bar(com + 0.2, profit, width = 0.4, label = "profit")
plt.legend()
```

* Now we see an excellent chart. For each company we have revenue and profit.

```python
# Next is horizontal bar chart. If we want to have bar chart in a horizontal fashion, so we use 'barh()' function.
plt.yticks(com, company)
plt.title("Companies Revenue and Profits")
plt.xlabel("revenue ($)")

plt.barh(com - 0.2, revenue, label = "revenue")
plt.barh(com + 0.2, profit, label = "profit")
plt.legend()
```
