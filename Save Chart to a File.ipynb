## Save chart to a File using saveFig
**SaveFig** in matplotlib is a function that allows the user to save their chart to a file. This can be useful for storing graphs or figures for future reference or to share with others. It allows the user to specify the file format, resolution, and other parameters for the saved chart.

```python
# Matplotlib Library:
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Let's again have some expenses values with addition to their labels.
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
# Let's create a pie chart from this two lists:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0,0.2,0,0])
plt.show()
```

```python
# Now to save the pie into a 'png, jpg, jpeg' file, so we call to savefig() function and in the bracket we supply the name
# of the file. 
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0,0.2,0,0])
plt.savefig("Pie_chart1.png", facecolor='white')
```

```python
# Now when we check the directory, the png file is created but it's not fit and is cutted. So if we use 'bbox_inches' attribute to be 'tight', it will produce the fit chart.
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0,0.2,0,0])
plt.savefig("Pie_chart2.png", facecolor='white', bbox_inches="tight", pad_inches=1)
```

```python
# To have some padding to the chart, we can use 'pad_inches' attribute:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0,0.2,0,0])
plt.savefig("Pie_chart3.png", facecolor='white', bbox_inches="tight", pad_inches=1.5)
```

```python
# To have transparent background, so we use tranparency property:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0,0.2,0,0])
plt.savefig("Pie_chart4.png", bbox_inches="tight", pad_inches=1.5, transparent = True)
```

```python
# You can also save chart as pdf file:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0,0.2,0,0])
plt.savefig("Pie_chart.pdf", bbox_inches="tight", pad_inches=1.5, transparent = True)
```

```python
# So first let's write the data in a list:
company = ["GOOGLE", "AMZN", "MSFT", "FB"]
revenue = [90, 136, 89, 27]
```

```python
# So next let creat a bar chart from the static data:
plt.title("US Technology Stocks")
plt.ylabel("Revenue(bln)")
plt.bar(company, revenue, label = "revenue")
plt.legend()
```

```python
# Now let's save the chart as jpg and pdf file:
plt.title("US Technology Stocks")
plt.ylabel("Revenue(bln)")
plt.bar(company, revenue, label = "revenue")
plt.legend()
plt.savefig("Bar_chart.jpg", facecolor='white', bbox_inches="tight", pad_inches=5)
plt.savefig("Bar_chart.pdf", facecolor='white', bbox_inches="tight", pad_inches=5)
```
