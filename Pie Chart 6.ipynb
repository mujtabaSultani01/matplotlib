## Pie chart in Matplotlib
A **pie chart** in Matplotlib is a type of graph that displays data in a circular graph divided into proportional slices, each representing a percentage of the whole. Pie charts are used to quickly and easily compare values between different categories. They are particularly useful when trying to show the relative proportions of different data points in a given set.

```python
# Matplotlib Library:
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Here we have two lists, first one is expenses values and the second one is expenses labels.
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
# To plot the pie plot of these two list, we simple call to plt 'pie()' method and pass the lists:
plt.pie(exp_vals,labels=exp_labels)
```

```python
# Now the chart is looking like oval or ellipse because of different pixesl for a-axis and y-axis. So we we define equal pixels for both the axis.
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
```

```python
# To remove the extra labels we use plt 'show()' method:
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

```python
# We can give shaddow to the chart:
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
# If we want to increase the size of pie chart, we can use radius argument:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius = 1.6) # Default the radius argument is 1.     
plt.show()
```

```python
# The next attribute is '' to display the percentage for each part of pie chart.
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius = 1.6, autopct='%0.2f%%') # Default the radius argument is 1.     
plt.show()
```

```python
# If we want to take out one of the specific pie, so we use 'explode' property:
# You specify the pie part by inserting the whole number of exe values.
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5,explode=[0,0,0.2,0,0])
plt.show()
```

```python
# 0.1 or 0.2 ... also specify the amount of going out the pie part:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5,explode=[0,0,0,0.3,0])
plt.show()
```

```python
# So by default the pie start from the first angle (Home Rent) and will go clock wise counter direction, if we want to start
# the pie chart with a different angle (45 degree), so we use 'startangle' argument:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5,explode=[0,0,0,0.3,0], startangle = 45)
plt.show()
```

```python
# Or to start with 180 degress:
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%0.2f%%',radius=1.5,explode=[0,0,0,0.3,0], startangle = 180)
plt.show()
```
